```python
import base64
```

```python
with open('../datasets/sample.webp', 'rb') as image_file:
    image_bytes = base64.b64encode(image_file.read())
```

```python
image_string = image_bytes.decode('utf-8')
```

```python
from PIL import Image
from io import BytesIO

image = Image.open(BytesIO(base64.b64decode(image_string)))
image
```
